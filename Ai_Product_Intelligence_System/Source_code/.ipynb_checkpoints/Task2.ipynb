**Workflow:** Image → CLIP Classification → Product Category → Recommendation Dictionary → Recommended Products

## 📚 Cell 2: Import Libraries

````python
import torch
import requests
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import networkx as nx
from PIL import Image
from io import BytesIO
from transformers import CLIPProcessor, CLIPModel
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully!")
print(f"🔧 PyTorch version: {torch.__version__}")
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"💻 Device: {device}")
````

## 🧠 Cell 3: Load CLIP Model

````python
print("⏳ Loading CLIP model (this may take a moment on first run)...")

model_name = "openai/clip-vit-base-patch32"
model = CLIPModel.from_pretrained(model_name).to(device)
processor = CLIPProcessor.from_pretrained(model_name)

model.eval()  # Set to evaluation mode
print("✅ CLIP model loaded successfully!")
print(f"📌 Model: {model_name}")
````

## 📖 Cell 4: Define Recommendation Dictionary & Categories

````python
# ─────────────────────────────────────────────────
# RECOMMENDATION DICTIONARY
# Key   → detected product category
# Value → list of complementary products
# ─────────────────────────────────────────────────
RECOMMENDATIONS = {
    "Running Shoe": [
        "Sports Socks",
        "Fitness Watch",
        "Water Bottle"
    ],
    "Laptop": [
        "Wireless Mouse",
        "Laptop Bag",
        "Keyboard"
    ],
    "Shirt": [
        "Jeans",
        "Shoes",
        "Watch"
    ],
    "Mobile": [
        "Phone Case",
        "Power Bank",
        "Earbuds"
    ]
}

# ─────────────────────────────────────────────────
# TEXT PROMPTS for CLIP zero-shot classification
# Each category gets a natural language description
# to help CLIP match images accurately
# ─────────────────────────────────────────────────
CATEGORY_PROMPTS = {
    "Running Shoe": "a photo of a running shoe or athletic sneaker",
    "Laptop":       "a photo of a laptop computer or notebook computer",
    "Shirt":        "a photo of a shirt, t-shirt, or top garment",
    "Mobile":       "a photo of a mobile phone or smartphone"
}

CATEGORIES   = list(CATEGORY_PROMPTS.keys())
CATEGORY_TEXT = list(CATEGORY_PROMPTS.values())

print("✅ Recommendation dictionary defined!")
print(f"📂 Supported categories: {CATEGORIES}")
````

## ⚙️ Cell 5: Core Functions

````python
# ─────────────────────────────────────────────────
# FUNCTION 1: Load image from URL or local file
# ─────────────────────────────────────────────────
def load_image(source: str) -> Image.Image:
    """
    Load a PIL image from a URL or a local file path.
    
    Args:
        source: URL string or local file path
    Returns:
        PIL Image in RGB format
    """
    if source.startswith("http://") or source.startswith("https://"):
        response = requests.get(source, timeout=10)
        response.raise_for_status()
        image = Image.open(BytesIO(response.content)).convert("RGB")
    else:
        image = Image.open(source).convert("RGB")
    return image


# ─────────────────────────────────────────────────
# FUNCTION 2: Classify image using CLIP
# ─────────────────────────────────────────────────
def classify_with_clip(image: Image.Image) -> tuple[str, dict]:
    """
    Run zero-shot CLIP classification against all category prompts.
    
    Args:
        image: PIL Image object
    Returns:
        (best_category, scores_dict) where scores_dict maps
        category → confidence percentage
    """
    inputs = processor(
        text=CATEGORY_TEXT,
        images=image,
        return_tensors="pt",
        padding=True
    ).to(device)

    with torch.no_grad():
        outputs      = model(**inputs)
        logits       = outputs.logits_per_image          # shape: [1, num_categories]
        probabilities = logits.softmax(dim=1).cpu().numpy()[0]  # normalise

    scores        = {cat: float(prob) * 100 for cat, prob in zip(CATEGORIES, probabilities)}
    best_category = max(scores, key=scores.get)
    return best_category, scores


# ─────────────────────────────────────────────────
# FUNCTION 3: Fetch recommendations from dictionary
# ─────────────────────────────────────────────────
def get_recommendations(category: str) -> list[str]:
    """
    Look up complementary products for the detected category.
    
    Args:
        category: one of the keys in RECOMMENDATIONS
    Returns:
        List of recommended product strings
    """
    return RECOMMENDATIONS.get(category, [])


print("✅ Core functions defined: load_image | classify_with_clip | get_recommendations")
````

## 📊 Cell 6: Visualization Functions

````python
# ─────────────────────────────────────────────────
# FUNCTION 4: Full dashboard visualization
# ─────────────────────────────────────────────────
def visualize_results(image: Image.Image,
                      category: str,
                      scores: dict,
                      recommendations: list[str]) -> None:
    """
    Render a 3-panel dashboard:
      Panel 1 – input image
      Panel 2 – CLIP confidence bar chart
      Panel 3 – recommendation graph
    """
    fig = plt.figure(figsize=(18, 6))
    fig.patch.set_facecolor('#0f0f1a')

    # ── Panel 1: Input Image ──────────────────────
    ax1 = fig.add_subplot(1, 3, 1)
    ax1.imshow(image)
    ax1.set_title("📷 Input Image", color='white', fontsize=13, fontweight='bold', pad=10)
    ax1.axis('off')
    ax1.set_facecolor('#1a1a2e')
    # Detected label below the image
    ax1.text(0.5, -0.04,
             f"Detected: {category}",
             transform=ax1.transAxes,
             ha='center', va='top',
             color='#00d4ff', fontsize=11, fontweight='bold')

    # ── Panel 2: Confidence Bar Chart ────────────
    ax2 = fig.add_subplot(1, 3, 2)
    ax2.set_facecolor('#1a1a2e')

    cats  = list(scores.keys())
    confs = list(scores.values())
    colors = ['#00d4ff' if c == category else '#4a4a6a' for c in cats]

    bars = ax2.barh(cats, confs, color=colors, edgecolor='none', height=0.5)
    for bar, conf in zip(bars, confs):
        ax2.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
                 f"{conf:.1f}%",
                 va='center', ha='left', color='white', fontsize=10)

    ax2.set_xlim(0, max(confs) * 1.25)
    ax2.set_xlabel('Confidence (%)', color='#aaaaaa', fontsize=10)
    ax2.set_title('🧠 CLIP Classification Scores', color='white',
                  fontsize=13, fontweight='bold', pad=10)
    ax2.tick_params(colors='#aaaaaa')
    ax2.spines[:].set_visible(False)
    for label in ax2.get_yticklabels():
        label.set_color('white')

    # ── Panel 3: Recommendation Graph ────────────
    ax3 = fig.add_subplot(1, 3, 3)
    ax3.set_facecolor('#1a1a2e')

    G = nx.DiGraph()
    G.add_node(category)
    for rec in recommendations:
        G.add_node(rec)
        G.add_edge(category, rec)

    # Star layout: centre = detected product
    pos = {category: (0, 0)}
    angles = np.linspace(0, 2 * np.pi, len(recommendations), endpoint=False)
    for rec, angle in zip(recommendations, angles):
        pos[rec] = (np.cos(angle) * 1.6, np.sin(angle) * 1.6)

    node_colors = ['#00d4ff' if n == category else '#ff6b6b' for n in G.nodes()]
    node_sizes  = [2200 if n == category else 1500 for n in G.nodes()]

    nx.draw_networkx_nodes(G, pos, ax=ax3, node_color=node_colors,
                           node_size=node_sizes, alpha=0.92)
    nx.draw_networkx_edges(G, pos, ax=ax3, edge_color='#555577',
                           arrows=True, arrowsize=18,
                           connectionstyle='arc3,rad=0.05',
                           min_source_margin=30, min_target_margin=30)

    # Wrap long labels
    labels = {n: n.replace(' ', '\n') for n in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels, ax=ax3,
                            font_size=8, font_color='white', font_weight='bold')

    ax3.set_title('🛍️ Recommended Products', color='white',
                  fontsize=13, fontweight='bold', pad=10)
    ax3.axis('off')

    # Legend
    legend_handles = [
        mpatches.Patch(color='#00d4ff', label='Detected Product'),
        mpatches.Patch(color='#ff6b6b', label='Recommended Product')
    ]
    ax3.legend(handles=legend_handles, loc='lower center',
               facecolor='#1a1a2e', edgecolor='none',
               labelcolor='white', fontsize=9,
               bbox_to_anchor=(0.5, -0.02))

    plt.suptitle('CLIP Product Recommendation System',
                 color='white', fontsize=16, fontweight='bold', y=1.01)
    plt.tight_layout()
    plt.savefig('recommendation_result.png', dpi=150,
                bbox_inches='tight', facecolor=fig.get_facecolor())
    plt.show()
    print("💾 Dashboard saved → recommendation_result.png")


# ─────────────────────────────────────────────────
# FUNCTION 5: ASCII tree (terminal-friendly)
# ─────────────────────────────────────────────────
def print_recommendation_tree(category: str, recommendations: list[str]) -> None:
    """
    Pretty-print the recommendation as an ASCII tree.
    """
    print()
    print("┌" + "─" * 40 + "┐")
    print(f"│  🎯 Detected Product: {category:<18}│")
    print("└" + "─" * 40 + "┘")
    print(f"        {category}")
    for i, rec in enumerate(recommendations):
        connector = "└────────►" if i == len(recommendations) - 1 else "├────────►"
        print(f"        {connector} {rec}")
    print()


print("✅ Visualization functions defined: visualize_results | print_recommendation_tree")
````

## 🚀 Cell 7: Main Pipeline Function

````python
def recommend_from_image(image_source: str, show_plot: bool = True) -> dict:
    """
    End-to-end pipeline:
      1. Load image from URL or local path
      2. Classify with CLIP
      3. Look up recommendations
      4. Visualize results

    Args:
        image_source: URL or local file path to the product image
        show_plot:    Whether to render the matplotlib dashboard

    Returns:
        dict with keys: 'category', 'confidence', 'all_scores', 'recommendations'
    """
    print("─" * 50)
    print("⏳ Step 1/4 – Loading image …")
    image = load_image(image_source)
    print(f"   Image size: {image.size[0]}×{image.size[1]} px")

    print("⏳ Step 2/4 – Running CLIP classification …")
    category, scores = classify_with_clip(image)
    confidence       = scores[category]
    print(f"   ✔ Category : {category}")
    print(f"   ✔ Confidence: {confidence:.1f}%")

    print("⏳ Step 3/4 – Fetching recommendations …")
    recs = get_recommendations(category)
    print(f"   ✔ Recommendations: {recs}")

    print("⏳ Step 4/4 – Rendering visualization …")
    print_recommendation_tree(category, recs)

    if show_plot:
        visualize_results(image, category, scores, recs)

    result = {
        "category":        category,
        "confidence":      round(confidence, 2),
        "all_scores":      {k: round(v, 2) for k, v in scores.items()},
        "recommendations": recs
    }
    print("\n✅ Pipeline complete!")
    print("─" * 50)
    return result


print("✅ Main pipeline function 'recommend_from_image' is ready!")
````

## 🔁 Cell 9: Batch Processing (all 4 categories)

````python
def batch_process(image_map: dict) -> dict:
    """
    Run the pipeline on multiple images and return all results.
    
    Args:
        image_map: dict of { label: image_url_or_path }
    Returns:
        dict of { label: result_dict }
    """
    all_results = {}
    for label, source in image_map.items():
        print(f"\n{'='*55}")
        print(f"  📸 Processing: {label}")
        print(f"{'='*55}")
        try:
            all_results[label] = recommend_from_image(source, show_plot=True)
        except Exception as e:
            print(f"  ❌ Error for '{label}': {e}")
            all_results[label] = {"error": str(e)}

    # Summary table
    print("\n" + "="*60)
    print("  📊 BATCH SUMMARY")
    print("="*60)
    print(f"  {'Image':<15} {'Detected':<15} {'Confidence':<12} {'Recommendations'}")
    print("  " + "-"*58)
    for label, res in all_results.items():
        if "error" not in res:
            rec_str = ", ".join(res['recommendations'])
            print(f"  {label:<15} {res['category']:<15} {res['confidence']:<12.1f}% {rec_str}")
        else:
            print(f"  {label:<15} {'ERROR':<15} {'-':<12} {res['error']}")
    print("="*60)
    return all_results


# Run batch
batch_results = batch_process(SAMPLE_IMAGES)
````

## ➕ Cell 11: Extend the Recommendation Dictionary

````python
# ─────────────────────────────────────────────────
# Add a new category at runtime — no restart needed
# ─────────────────────────────────────────────────

def add_category(name: str, description_prompt: str, recs: list[str]) -> None:
    """
    Dynamically add a new product category to the system.
    
    Args:
        name:                Category name (must be unique)
        description_prompt:  Natural language prompt for CLIP
        recs:                List of recommended companion products
    """
    if name in RECOMMENDATIONS:
        print(f"⚠️  '{name}' already exists. Updating…")

    RECOMMENDATIONS[name]       = recs
    CATEGORY_PROMPTS[name]      = description_prompt
    CATEGORIES.append(name)
    CATEGORY_TEXT.append(description_prompt)

    print(f"✅ Category '{name}' added!")
    print(f"   Prompt  : {description_prompt}")
    print(f"   Recs    : {recs}")
    print(f"   Total categories now: {len(CATEGORIES)}")


# ── Example: add Camera category ─────────────────
add_category(
    name="Camera",
    description_prompt="a photo of a digital camera or DSLR camera",
    recs=["Camera Bag", "Memory Card", "Tripod"]
)

print("\n📋 All categories now available:")
for cat, recs in RECOMMENDATIONS.items():
    print(f"  • {cat} → {recs}")
````

# 🗂️ Phase 4 — Task 2: Unique Product Catalog
**Workflow:** embeddings.pkl → Cosine Similarity → Clustering → Representative Selection → Final Catalog

## 📦 Cell 13: Install Extra Dependencies & Load Embeddings

````python
# ─────────────────────────────────────────────────────────────────────
# Install extra libs needed for Phase 4 Task 2
# (torch / transformers / pillow already installed in Cell 1)
# ─────────────────────────────────────────────────────────────────────
!pip install scikit-learn umap-learn --quiet
print("✅ scikit-learn & umap-learn installed")

# ── Imports ───────────────────────────────────────────────────────────
import os
import pickle
import numpy as np
import torch
import requests
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
from PIL import Image
from io import BytesIO
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# ── Attempt to import UMAP (optional — falls back to PCA) ────────────
try:
    from umap import UMAP
    UMAP_AVAILABLE = True
    print("✅ UMAP available")
except ImportError:
    UMAP_AVAILABLE = False
    print("⚠️  UMAP not available — PCA will be used for 2D projection")

# ──────────────────────────────────────────────────────────────────────
# STEP 13 — Load embeddings.pkl
#
# Expected format (either is accepted):
#   Format A — dict : { product_id: np.ndarray(shape=[D]) }
#   Format B — dict : { 'ids': [...], 'embeddings': np.ndarray([N, D]) }
#
# If the file is absent we GENERATE a synthetic dataset so every
# downstream cell runs end-to-end without requiring real data.
# ──────────────────────────────────────────────────────────────────────
EMBEDDINGS_PATH = "embeddings.pkl"   # ← change to your real path
SYNTHETIC_MODE  = False

def _make_synthetic_embeddings(n_products: int = 30,
                               dim: int = 512,
                               n_clusters: int = 6,
                               seed: int = 42) -> tuple[list[str], np.ndarray]:
    """
    Generate realistic synthetic embeddings:
      • n_clusters tight groups (duplicates within each group)
      • remaining products are singletons (unique)
    """
    rng = np.random.default_rng(seed)
    centres = rng.standard_normal((n_clusters, dim))
    centres = centres / np.linalg.norm(centres, axis=1, keepdims=True)

    product_names = [
        "Blue Shirt A",  "Blue Shirt B",  "Blue Shirt C",    # cluster 0
        "Running Shoe A","Running Shoe B","Running Shoe C",  # cluster 1
        "Laptop A",      "Laptop B",       "Laptop C",        # cluster 2
        "Watch A",       "Watch B",                           # cluster 3
        "Jeans A",       "Jeans B",                           # cluster 4
        "Phone Case A",  "Phone Case B",                      # cluster 5
        "Water Bottle",  "Fitness Watch",  "Earbuds",         # singletons
        "Keyboard",      "Laptop Bag",     "Sports Socks",
        "Power Bank",    "Wireless Mouse", "Tripod",
        "Memory Card",   "Camera Bag",     "Shoes",
        "Jeans",         "Watch",          "Phone Case"
    ]
    cluster_map = {
        "Blue Shirt A": 0,  "Blue Shirt B": 0,  "Blue Shirt C": 0,
        "Running Shoe A": 1,"Running Shoe B": 1,"Running Shoe C": 1,
        "Laptop A": 2,      "Laptop B": 2,       "Laptop C": 2,
        "Watch A": 3,       "Watch B": 3,
        "Jeans A": 4,       "Jeans B": 4,
        "Phone Case A": 5,  "Phone Case B": 5,
    }
    embeddings = []
    for name in product_names:
        if name in cluster_map:
            cid = cluster_map[name]
            noise = rng.standard_normal(dim) * 0.02   # very small noise → high similarity
            vec = centres[cid] + noise
        else:
            vec = rng.standard_normal(dim)             # random → unique
        embeddings.append(vec)

    emb_array = np.array(embeddings, dtype=np.float32)
    emb_array = emb_array / np.linalg.norm(emb_array, axis=1, keepdims=True)  # L2-normalise
    return product_names, emb_array


# ── Try loading the real file; fall back to synthetic ────────────────
if os.path.exists(EMBEDDINGS_PATH):
    with open(EMBEDDINGS_PATH, "rb") as f:
        raw = pickle.load(f)

    if isinstance(raw, dict) and 'embeddings' in raw:
        # Format B
        product_ids  = list(raw['ids'])
        emb_matrix   = np.array(raw['embeddings'], dtype=np.float32)
    elif isinstance(raw, dict):
        # Format A
        product_ids  = list(raw.keys())
        emb_matrix   = np.array(list(raw.values()), dtype=np.float32)
    else:
        raise ValueError("Unrecognised embeddings.pkl format")

    # L2-normalise so cosine sim == dot product
    emb_matrix = emb_matrix / (np.linalg.norm(emb_matrix, axis=1, keepdims=True) + 1e-10)
    print(f"✅ Loaded {len(product_ids)} embeddings from '{EMBEDDINGS_PATH}'")
    print(f"   Embedding dim : {emb_matrix.shape[1]}")

else:
    SYNTHETIC_MODE = True
    print(f"⚠️  '{EMBEDDINGS_PATH}' not found — running in SYNTHETIC MODE")
    product_ids, emb_matrix = _make_synthetic_embeddings()
    print(f"✅ Synthetic dataset: {len(product_ids)} products, dim={emb_matrix.shape[1]}")
    print()
    print("📋 Products loaded:")
    for pid in product_ids:
        print(f"   • {pid}")
````

## 🔢 Cell 14: Calculate Cosine Similarity & Flag Duplicates

````python
# ──────────────────────────────────────────────────────────────────────
# STEP 14 — Cosine Similarity Matrix
#
# Because embeddings are L2-normalised, cosine_similarity == dot product
# Threshold: similarity > 0.95 → treated as duplicate pair
# ──────────────────────────────────────────────────────────────────────
SIMILARITY_THRESHOLD = 0.95   # ← tune this (lower = stricter; higher = more duplicates)

# Full N×N cosine similarity matrix
sim_matrix = cosine_similarity(emb_matrix)          # shape: [N, N]
N          = len(product_ids)

# ── Identify all duplicate pairs ─────────────────────────────────────
duplicate_pairs = []
for i in range(N):
    for j in range(i + 1, N):
        if sim_matrix[i, j] > SIMILARITY_THRESHOLD:
            duplicate_pairs.append((product_ids[i], product_ids[j], sim_matrix[i, j]))

# ── Pretty-print results ─────────────────────────────────────────────
print(f"📐 Similarity matrix shape : {sim_matrix.shape}")
print(f"⚙️  Threshold               : > {SIMILARITY_THRESHOLD}")
print(f"🔍 Duplicate pairs found   : {len(duplicate_pairs)}")
print()

if duplicate_pairs:
    print(f"{'Product A':<22} {'Product B':<22} {'Similarity':>10}")
    print("─" * 58)
    for a, b, score in sorted(duplicate_pairs, key=lambda x: -x[2]):
        status = "🔴 DUPLICATE" if score > SIMILARITY_THRESHOLD else ""
        print(f"{a:<22} {b:<22} {score:>10.4f}  {status}")
else:
    print("✅ No duplicates found at this threshold.")

# ── Heatmap of top-20 × top-20 similarity scores ────────────────────
top_n    = min(20, N)
sub_sim  = sim_matrix[:top_n, :top_n]
sub_ids  = product_ids[:top_n]

fig, ax = plt.subplots(figsize=(11, 9))
fig.patch.set_facecolor('#0f0f1a')
ax.set_facecolor('#0f0f1a')

im = ax.imshow(sub_sim, cmap='RdYlGn', vmin=0, vmax=1, aspect='auto')

# Overlay similarity values
for i in range(top_n):
    for j in range(top_n):
        val = sub_sim[i, j]
        color = 'black' if val > 0.6 else 'white'
        ax.text(j, i, f"{val:.2f}", ha='center', va='center',
                fontsize=6, color=color, fontweight='bold')

ax.set_xticks(range(top_n))
ax.set_yticks(range(top_n))
ax.set_xticklabels(sub_ids, rotation=45, ha='right', color='white', fontsize=8)
ax.set_yticklabels(sub_ids, color='white', fontsize=8)

cbar = plt.colorbar(im, ax=ax, fraction=0.03, pad=0.04)
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(cbar.ax.yaxis.get_ticklabels(), color='white')
cbar.set_label('Cosine Similarity', color='white', fontsize=10)

# Highlight duplicate pairs with red boxes
for a, b, score in duplicate_pairs:
    if a in sub_ids and b in sub_ids:
        xi, yi = sub_ids.index(a), sub_ids.index(b)
        for (r, c) in [(yi, xi), (xi, yi)]:
            rect = plt.Rectangle((c - 0.5, r - 0.5), 1, 1,
                                  linewidth=2, edgecolor='red', facecolor='none')
            ax.add_patch(rect)

ax.set_title(f"Step 14 — Cosine Similarity Heatmap  (threshold = {SIMILARITY_THRESHOLD})",
             color='white', fontsize=13, fontweight='bold', pad=12)
plt.tight_layout()
plt.savefig('step14_similarity_heatmap.png', dpi=150,
            bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
print("💾 Saved → step14_similarity_heatmap.png")
````

## 🔵 Cell 15: Cluster Similar Images (DBSCAN + Agglomerative)

````python
# ──────────────────────────────────────────────────────────────────────
# STEP 15 — Cluster Similar Images
#
# Strategy A: DBSCAN on cosine distance matrix
#   • eps       = 1 - SIMILARITY_THRESHOLD  (distance threshold)
#   • min_samples = 1  so every point belongs to a cluster
#   • Noise label (-1) → unique singleton treated as its own cluster
#
# Strategy B: Agglomerative Clustering (Ward linkage)
#   • distance_threshold converts the similarity threshold to distance
#   • n_clusters=None so the threshold controls the cut
# ──────────────────────────────────────────────────────────────────────

CLUSTERING_METHOD = "DBSCAN"   # ← swap to "Agglomerative" if preferred

# Convert cosine similarity matrix → cosine distance matrix
dist_matrix = 1.0 - sim_matrix
np.fill_diagonal(dist_matrix, 0.0)          # numerical safety
dist_matrix = np.clip(dist_matrix, 0, None)  # no negative distances

eps = 1.0 - SIMILARITY_THRESHOLD            # distance equivalent of the threshold

if CLUSTERING_METHOD == "DBSCAN":
    clusterer = DBSCAN(eps=eps, min_samples=1, metric='precomputed')
    raw_labels = clusterer.fit_predict(dist_matrix)

    # DBSCAN labels noise as -1 → give each noise point a unique cluster id
    labels = raw_labels.copy()
    next_id = labels.max() + 1
    for i, lbl in enumerate(labels):
        if lbl == -1:
            labels[i] = next_id
            next_id += 1

    noise_count = (raw_labels == -1).sum()
    print(f"🔵 DBSCAN  eps={eps:.3f}  |  noise points reassigned as singletons: {noise_count}")

else:  # Agglomerative
    clusterer = AgglomerativeClustering(
        n_clusters=None,
        metric='precomputed',
        linkage='complete',
        distance_threshold=eps
    )
    labels = clusterer.fit_predict(dist_matrix)
    print(f"🔵 Agglomerative  distance_threshold={eps:.3f}")

n_clusters = len(set(labels))
print(f"📊 Total clusters formed  : {n_clusters}")
print(f"📦 Original product count : {N}")
print(f"📉 Reduction              : {N - n_clusters} duplicates removed")
print()

# ── Build cluster → members mapping ──────────────────────────────────
from collections import defaultdict
clusters: dict[int, list[str]] = defaultdict(list)
for pid, lbl in zip(product_ids, labels):
    clusters[int(lbl)].append(pid)

# ── Print non-singleton clusters (the real duplicates) ───────────────
dup_clusters = {k: v for k, v in clusters.items() if len(v) > 1}
print(f"🔴 Duplicate clusters ({len(dup_clusters)} found):")
for cid, members in sorted(dup_clusters.items()):
    print(f"\n  Cluster {cid}:")
    for m in members:
        print(f"    ├── {m}")

# Singletons
sing_clusters = {k: v for k, v in clusters.items() if len(v) == 1}
print(f"\n🟢 Unique singletons : {len(sing_clusters)}")
for _, members in sorted(sing_clusters.items()):
    print(f"   • {members[0]}")
````

## 🏆 Cell 16: Select Representative Product per Cluster

````python
# ──────────────────────────────────────────────────────────────────────
# STEP 16 — Select Representative Product
#
# Strategy: centroid-based selection
#   1. Compute the mean embedding of all members in the cluster
#   2. Pick the member whose embedding is closest to the centroid
#      (highest cosine similarity to mean) → most "average" product
#
# This is more principled than simply taking the first member because
# the centroid representative is the one that best characterises the
# whole cluster visually.
# ──────────────────────────────────────────────────────────────────────

pid_to_idx = {pid: i for i, pid in enumerate(product_ids)}

def select_representative(cluster_members: list[str]) -> tuple[str, list[str]]:
    """
    Choose the centroid-nearest member as the cluster representative.

    Returns:
        (representative, duplicates_to_remove)
    """
    if len(cluster_members) == 1:
        return cluster_members[0], []

    idxs      = [pid_to_idx[m] for m in cluster_members]
    vecs      = emb_matrix[idxs]               # [k, D]
    centroid  = vecs.mean(axis=0, keepdims=True)  # [1, D]
    centroid  = centroid / (np.linalg.norm(centroid) + 1e-10)
    sims      = cosine_similarity(centroid, vecs)[0]  # [k]
    best_idx  = int(np.argmax(sims))
    rep       = cluster_members[best_idx]
    dups      = [m for m in cluster_members if m != rep]
    return rep, dups


catalog        : list[str] = []    # kept products
removed        : list[str] = []    # removed duplicates
cluster_report : list[dict] = []   # structured report per cluster

for cid, members in sorted(clusters.items()):
    rep, dups = select_representative(members)
    catalog.append(rep)
    removed.extend(dups)
    cluster_report.append({
        'cluster_id':      cid,
        'members':         members,
        'representative':  rep,
        'duplicates':      dups
    })

# ── Print decision log ────────────────────────────────────────────────
print("="*60)
print("  Step 16 — Representative Selection Report")
print("="*60)
for r in cluster_report:
    if r['duplicates']:      # only print clusters with actual duplicates
        print(f"\n  Cluster {r['cluster_id']}  ({len(r['members'])} members)")
        print(f"  {'Member':<25} {'Decision'}")
        print("  " + "-"*40)
        for m in r['members']:
            tag = "✅ KEEP (representative)" if m == r['representative'] else "❌ REMOVE (duplicate)"
            print(f"  {m:<25} {tag}")

print()
print("─"*60)
print(f"  Original count   : {N}")
print(f"  Kept (catalog)   : {len(catalog)}")
print(f"  Removed (dups)   : {len(removed)}")
print("─"*60)
````

## 📋 Cell 17: Generate Final Unique Catalog

````python
# ──────────────────────────────────────────────────────────────────────
# STEP 17 — Generate Final Catalog
#
# Outputs:
#   • catalog (list)          — unique product names / IDs
#   • final_catalog.pkl       — serialised catalog for downstream use
#   • final_catalog.txt       — human-readable plain text list
# ──────────────────────────────────────────────────────────────────────

# ── 1. Build catalog dict (id → embedding) ───────────────────────────
catalog_embeddings = {
    pid: emb_matrix[pid_to_idx[pid]]
    for pid in catalog
}

# ── 2. Save catalog as pickle ─────────────────────────────────────────
catalog_payload = {
    'ids':         catalog,
    'embeddings':  np.stack([catalog_embeddings[pid] for pid in catalog]),
    'removed':     removed,
    'cluster_report': cluster_report
}
with open("final_catalog.pkl", "wb") as f:
    pickle.dump(catalog_payload, f)
print("💾 Saved → final_catalog.pkl")

# ── 3. Save human-readable text list ─────────────────────────────────
with open("final_catalog.txt", "w") as f:
    f.write("UNIQUE PRODUCT CATALOG\n")
    f.write("=" * 30 + "\n")
    for i, pid in enumerate(catalog, 1):
        f.write(f"{i:>3}. {pid}\n")
    f.write("\n" + "=" * 30 + "\n")
    f.write(f"Total unique products: {len(catalog)}\n")
    f.write(f"Duplicates removed   : {len(removed)}\n")
print("💾 Saved → final_catalog.txt")

# ── 4. Print final catalog ────────────────────────────────────────────
print()
print("╔" + "═"*40 + "╗")
print("║      ✅  FINAL UNIQUE CATALOG             ║")
print("╠" + "═"*40 + "╣")
for i, pid in enumerate(catalog, 1):
    print(f"║  {i:>2}. {pid:<35}║")
print("╠" + "═"*40 + "╣")
print(f"║  Total unique : {len(catalog):<24}║")
print(f"║  Removed dups : {len(removed):<24}║")
print("╚" + "═"*40 + "╝")

# ── 5. Show removed duplicates ───────────────────────────────────────
if removed:
    print()
    print("🗑️  Products removed as duplicates:")
    for r in removed:
        print(f"   ✗ {r}")
````

## 📊 Cell 18: Full Pipeline Visualization Dashboard

````python
# ──────────────────────────────────────────────────────────────────────
# STEP 18 — 4-Panel Visualization
#
# Panel 1 — Original Dataset  (all products, coloured by cluster)
# Panel 2 — Embedding Space   (2-D projection via UMAP or PCA)
# Panel 3 — Clustering        (cluster boundaries + duplicates marked)
# Panel 4 — Unique Catalog    (final selected representatives)
# ──────────────────────────────────────────────────────────────────────

# ── 2-D projection ───────────────────────────────────────────────────
print("⏳ Computing 2-D projection …")
if UMAP_AVAILABLE and N >= 10:
    reducer = UMAP(n_components=2, random_state=42,
                   n_neighbors=min(10, N-1), min_dist=0.1)
    method_label = "UMAP"
else:
    reducer = PCA(n_components=2, random_state=42)
    method_label = "PCA"
coords_2d = reducer.fit_transform(emb_matrix)   # [N, 2]
print(f"✅ {method_label} projection ready")

# ── Colour palette (one colour per cluster) ───────────────────────────
unique_labels  = sorted(set(labels))
cmap           = plt.cm.get_cmap('tab20', max(len(unique_labels), 1))
label_to_color = {lbl: cmap(i) for i, lbl in enumerate(unique_labels)}
point_colors   = [label_to_color[lbl] for lbl in labels]

rep_set        = set(catalog)
is_rep         = [pid in rep_set for pid in product_ids]
is_dup         = [pid in set(removed) for pid in product_ids]

# ── Figure setup ─────────────────────────────────────────────────────
fig = plt.figure(figsize=(22, 18))
fig.patch.set_facecolor('#0d0d1a')
gs  = gridspec.GridSpec(2, 2, figure=fig, hspace=0.38, wspace=0.3)

TITLE_STYLE = dict(color='white', fontsize=13, fontweight='bold', pad=12)
TICK_STYLE  = dict(colors='#888888')
SPINE_COLOR = '#333355'

def style_ax(ax, title):
    ax.set_facecolor('#13132a')
    ax.set_title(title, **TITLE_STYLE)
    ax.tick_params(**TICK_STYLE)
    for spine in ax.spines.values():
        spine.set_edgecolor(SPINE_COLOR)

# ────────────────────────────────────────────────────────────────────
# Panel 1 — Original Dataset
# ────────────────────────────────────────────────────────────────────
ax1 = fig.add_subplot(gs[0, 0])
style_ax(ax1, f"① Original Dataset  ({N} products)")

ax1.scatter(coords_2d[:, 0], coords_2d[:, 1],
            c=point_colors, s=90, alpha=0.85, edgecolors='white', linewidths=0.4)

for i, (pid, (x, y)) in enumerate(zip(product_ids, coords_2d)):
    ax1.annotate(pid, (x, y), fontsize=6, color='#cccccc',
                 xytext=(4, 4), textcoords='offset points')

ax1.set_xlabel(f"{method_label} dim-1", color='#888888', fontsize=9)
ax1.set_ylabel(f"{method_label} dim-2", color='#888888', fontsize=9)

# ────────────────────────────────────────────────────────────────────
# Panel 2 — Embedding Space (coloured by cluster, no labels)
# ────────────────────────────────────────────────────────────────────
ax2 = fig.add_subplot(gs[0, 1])
style_ax(ax2, f"② Embedding Space  ({method_label} projection)")

for lbl in unique_labels:
    mask = [l == lbl for l in labels]
    xs   = coords_2d[mask, 0]
    ys   = coords_2d[mask, 1]
    ax2.scatter(xs, ys, c=[label_to_color[lbl]] * xs.shape[0],
                s=100, alpha=0.88, edgecolors='white', linewidths=0.4,
                label=f"Cluster {lbl}")

ax2.set_xlabel(f"{method_label} dim-1", color='#888888', fontsize=9)
ax2.set_ylabel(f"{method_label} dim-2", color='#888888', fontsize=9)
if n_clusters <= 12:
    leg = ax2.legend(fontsize=7, framealpha=0.2, labelcolor='white',
                     facecolor='#1a1a2e', edgecolor='none',
                     loc='best', ncol=2)

# ────────────────────────────────────────────────────────────────────
# Panel 3 — Clustering (representatives ★, duplicates ✗)
# ────────────────────────────────────────────────────────────────────
ax3 = fig.add_subplot(gs[1, 0])
style_ax(ax3, f"③ Clustering  (method: {CLUSTERING_METHOD})")

# Background scatter (all points, muted)
ax3.scatter(coords_2d[:, 0], coords_2d[:, 1],
            c=point_colors, s=70, alpha=0.35, edgecolors='none')

# Representatives — large gold stars
rep_mask = np.array(is_rep)
ax3.scatter(coords_2d[rep_mask, 0], coords_2d[rep_mask, 1],
            marker='*', s=280, c='gold', edgecolors='white',
            linewidths=0.6, zorder=5, label='Representative')

# Duplicates — red X
dup_mask = np.array(is_dup)
ax3.scatter(coords_2d[dup_mask, 0], coords_2d[dup_mask, 1],
            marker='X', s=120, c='#ff4444', edgecolors='white',
            linewidths=0.5, zorder=4, label='Duplicate (removed)')

# Draw arrows from each duplicate to its representative
for r in cluster_report:
    if not r['duplicates']:
        continue
    rep_idx = pid_to_idx[r['representative']]
    rx, ry  = coords_2d[rep_idx]
    for dup in r['duplicates']:
        di      = pid_to_idx[dup]
        dx, dy  = coords_2d[di]
        ax3.annotate("", xy=(rx, ry), xytext=(dx, dy),
                     arrowprops=dict(arrowstyle="->", color='#ff4444',
                                     lw=1.2, alpha=0.7))

# Annotate representatives
for i, pid in enumerate(product_ids):
    if is_rep[i]:
        ax3.annotate(pid, (coords_2d[i, 0], coords_2d[i, 1]),
                     fontsize=6.5, color='gold', fontweight='bold',
                     xytext=(5, 5), textcoords='offset points')

ax3.set_xlabel(f"{method_label} dim-1", color='#888888', fontsize=9)
ax3.set_ylabel(f"{method_label} dim-2", color='#888888', fontsize=9)
ax3.legend(fontsize=8, framealpha=0.25, labelcolor='white',
           facecolor='#1a1a2e', edgecolor='none', loc='best')

# ────────────────────────────────────────────────────────────────────
# Panel 4 — Unique Catalog (bar chart: before vs after)
# ────────────────────────────────────────────────────────────────────
ax4 = fig.add_subplot(gs[1, 1])
style_ax(ax4, "④ Unique Catalog  (before vs after deduplication)")

# Horizontal grouped bar: original count vs catalog count per cluster
dup_clust_ids   = sorted(dup_clusters.keys())
dup_clust_names = [', '.join(clusters[cid]) for cid in dup_clust_ids]
orig_sizes      = [len(clusters[cid]) for cid in dup_clust_ids]
final_sizes     = [1] * len(dup_clust_ids)   # always 1 after dedup

y_pos = np.arange(len(dup_clust_ids))
bar_h = 0.35

bars_orig  = ax4.barh(y_pos - bar_h/2, orig_sizes,  bar_h,
                      color='#ff6b6b', label='Before (duplicates)', alpha=0.85)
bars_final = ax4.barh(y_pos + bar_h/2, final_sizes, bar_h,
                      color='#00d4ff', label='After  (unique)',      alpha=0.85)

# Labels inside bars
for bar, val in zip(bars_orig, orig_sizes):
    ax4.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
             str(val), va='center', color='#ff6b6b', fontsize=10, fontweight='bold')
for bar in bars_final:
    ax4.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
             '1', va='center', color='#00d4ff', fontsize=10, fontweight='bold')

# Tick labels: show only the representative name
rep_names = [clusters[cid][[pid_to_idx.get(m, 0) for m in clusters[cid]].index(
             min([pid_to_idx.get(m, 0) for m in clusters[cid]]))] for cid in dup_clust_ids]

# Simpler: use the representative from cluster_report
rep_by_cid = {r['cluster_id']: r['representative'] for r in cluster_report}
tick_labels = [rep_by_cid.get(cid, str(cid)) for cid in dup_clust_ids]

ax4.set_yticks(y_pos)
ax4.set_yticklabels(tick_labels, color='white', fontsize=9)
ax4.set_xlabel('Product Count', color='#888888', fontsize=9)
ax4.set_xlim(0, max(orig_sizes) * 1.3 if orig_sizes else 4)
ax4.legend(fontsize=9, framealpha=0.25, labelcolor='white',
           facecolor='#1a1a2e', edgecolor='none', loc='lower right')

# Totals annotation
ax4.text(0.97, 0.02,
         f"Original: {N}  →  Catalog: {len(catalog)}  (−{len(removed)} dups)",
         transform=ax4.transAxes, ha='right', va='bottom',
         color='white', fontsize=9,
         bbox=dict(boxstyle='round,pad=0.4', facecolor='#1a1a2e', alpha=0.7))

# ── Super-title & save ────────────────────────────────────────────────
fig.suptitle(
    "Phase 4 · Task 2 — Unique Product Catalog Pipeline\n"
    "Original Dataset  →  Embedding  →  Clustering  →  Unique Catalog",
    color='white', fontsize=15, fontweight='bold', y=1.01
)
plt.savefig('step18_catalog_pipeline.png', dpi=150,
            bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
print("💾 Dashboard saved → step18_catalog_pipeline.png")

# ── Final summary ─────────────────────────────────────────────────────
print()
print("="*55)
print("  🏁  PIPELINE COMPLETE — SUMMARY")
print("="*55)
print(f"  Clustering method   : {CLUSTERING_METHOD}")
print(f"  Similarity threshold: {SIMILARITY_THRESHOLD}")
print(f"  Projection method   : {method_label}")
print(f"  Original products   : {N}")
print(f"  Duplicate clusters  : {len(dup_clusters)}")
print(f"  Duplicates removed  : {len(removed)}")
print(f"  Unique catalog size : {len(catalog)}")
print("="*55)
print()
print("📂 Output files:")
print("   • final_catalog.pkl            — catalog embeddings + metadata")
print("   • final_catalog.txt            — plain text product list")
print("   • step14_similarity_heatmap.png")
print("   • step18_catalog_pipeline.png")
````
